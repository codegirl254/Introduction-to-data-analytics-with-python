```python
q1 = df["Sales"].quantile(0.25)
q3 = df["Sales"].quantile(0.75)
iqr = q3-q1
print(q1,q3,iqr)
```

```python
lower_bond =q1-1.5*iqr
lower_bond
```

```python
upper_bond=q3+1.5*iqr
upper_bond
```

```python
required = df[(df["Sales"]>= lower_bond)& (df["Sales"]<=upper_bond)]
print("All sales" ,df["Sales"].values)
print("Sales after removing outliers",required["Sales"].values)
```
