```python
"""
Suppose you have a dataset containing information about car sales. The dataset has the 
following columns: "Manufacturer", "Model", "Year", "Sales", and "Revenue". Using Pandas, 
perform the following tasks:

"""

import pandas as pd
```

```python
 # Load the dataset from a CSV file named "car_sales.csv" into a DataFrame called df.

filepath ="Data_Files/car_sales.csv"
df=pd.read_csv(filepath)
```

```python
df
```

```python
# Display the last five rows of the DataFrame to inspect the data.
df.tail()
```

```python
# Calculate and display the total revenue generated by each manufacturer.
df.groupby("Manufacturer")["Revenue"].sum()
```

```python
# Filter the data to select car models sold in the year 2022.
df[df["Year"]==2022]
```

```python
"""
Add a new column named "Profit" to the DataFrame, where the profit is calculated as 
10% of the revenue.
"""

df["Profit"]=df["Revenue"]*(10/100)
df
```

```python
 # Group the data by year and calculate the total sales for each year.
df.groupby("Year")["Sales"].sum()
```
